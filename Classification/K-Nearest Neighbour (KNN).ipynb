```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing dataset
data = pd.read_csv("Social_Network_Ads.csv") 
```

```python
# checking top 5 rows in data
data.head()
```

```python
# Information about the data
data.info()
```

```python
#check for missing values
data.isnull().sum() 
```

```python
# partitioning the data into features and labels as X and y respectively
X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
X
```

```python
y
```

```python
# Splitting the Dataset 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 5)
```

```python
X_train
```

```python
y_train
```

There is no need of Encoder because data contains only numerical values not String.

```python
# Scaling the features
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
# Training the model
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

```python
# Prediction on new sample data
classifier.predict(sc.transform([[56,22000]]))
```

```python
# prediction on test data
pred = classifier.predict(X_test)
pred
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score, r2_score
cm = confusion_matrix(y_test, pred)
print(cm)
print("Accuracy - ", accuracy_score(y_test, pred))
print("R2 Score - ", r2_score(y_test, pred))
```

Here, R2 score is better than the logistic regression and SVM.
